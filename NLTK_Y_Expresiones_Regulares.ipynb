```
import nltk
nltk.download('cess_esp')
```

```
import re
corpus = nltk.corpus.cess_esp.sents()
print(corpus)
print(len(corpus))
```

```
flatten = [w for l in corpus for w in l]
print (len(flatten))
```

```
print(flatten[:200])
```

```
arr = [w for w in flatten if re.search('es',w)]

```

```
arr = [w for w in flatten if re.search('es$',w)]
print(arr[:5])
```

```
arr = [w for w in flatten if re.search('^es',w)]
print(arr[:5])
```

```
#Rango [a-z] [ghi]
arr = [w for w in flatten if re.search('^[ghi]',w)]
print(arr[:10])
```

```
#Rango [a-c] 
arr = [w for w in flatten if re.search('^[a-c]',w)]
print(arr[:10])
```

```
#clausura
# * repetir 0 o mas veces
# +  repetir 1 o mas veces
# pruebo con * 
arr = [w for w in flatten if re.search('^(no)*',w)]
print(arr[:10])
# imprime todas porque en * el minimo es 0, pero en +, el minimo es 1
```

```
#clausura
# * repetir 0 o mas veces
# +  repetir 1 o mas veces
arr = [w for w in flatten if re.search('^(no)+',w)]
print(arr[:10])
```

```
#texto normal
print('esta es \n una prueba')
#texto en crudo , o texto plano , o raw, se ponr 'r' antes del string
print(r'esta es \n una prueba')
```

```
texto = """ Cuando sea el rey del mundo  (imaginaba él en su cabeza) no tendré que  preocuparme por estas bobadas. 
            Era solo un niño de 7 años, pero pensaba que podría ser cualquier cosa que su imaginación le permitiera visualizar en su cabeza ..."""
```

```
#caso 1 : tokenizar por espacios vacios
print(re.split(r' ',texto))
```

```
#caso 2 : tokenizacion usando expresiones regulares
print(re.split(r'[ \t\n]+',texto))
```

```
#caso 3 : tokenizacion usando expresiones regulares (MODIFICADO)
print(re.split(r'[ \t\n\W]+',texto))
#se eliminaran los parentesis
```

```
texto = 'En los E.U. esa postal vale $15.50 ...'
print(re.split(r'[ \t\n\W]+',texto))
```

```
pattern = r'''(?x)                 # set flag to allow verbose regexps
              (?:[A-Z]\.)+         # abbreviations, e.g. U.S.A.
              | \w+(?:-\w+)*       # words with optional internal hyphens
              | \$?\d+(?:\.\d+)?%? # currency and percentages, e.g. $12.40, 82%
              | \.\.\.             # ellipsis
              | [][.,;"'?():-_`]   # these are separate tokens; includes ], [
'''
```

```
nltk.regexp_tokenize(texto, pattern)
```
